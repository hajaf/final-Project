```python
import numpy as np 
import pandas as pd 
import os
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in csv
df_2015 = pd.read_csv('Resources/cleaned_2015.csv')  
df_2016 = pd.read_csv('Resources/cleaned_2016.csv')  
```

```python
df_2015.head()
```

```python
#List of top 50 country and Freedom of 2015 
sns.set_theme(style='dark')
plt.figure(figsize=(24,12))
ax=sns.scatterplot(data=df_2015.head(50),x='Country',y='Freedom',size='Score',sizes=(500,2000)
                   ,hue='Score')
plt.xticks(rotation=90,fontsize=20,color='midnightblue')
plt.yticks(fontsize=15,color='midnightblue')
plt.xlabel('Country or region',size=20,color='midnightblue')
plt.ylabel('Freedom',size=20,color='midnightblue')
plt.show()
```

```python
#List of bottom 50 country and Freedom of 2015 
sns.set_theme(style='dark')
plt.figure(figsize=(24,12))
ax=sns.scatterplot(data=df_2015.tail(50),x='Country',y='Freedom',size='Score',sizes=(500,2000)
                  ,hue='Score')
plt.xticks(rotation=90,fontsize=20,color='midnightblue')
plt.yticks(fontsize=20,color='midnightblue')
plt.xlabel('Country',size=20,color='midnightblue')
plt.ylabel('Freedom')
plt.show()
```
